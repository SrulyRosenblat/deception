````python
{'name': 'Streaming platform',
'home': 'A streaming platform home page',
'pricing': 'A streaming platform pricing page',
'checkout_flow': 'A streaming platform checkout flow'},
import os
def generate_site(idea,context= None,model="openai/gpt-4o-mini"):
    name= idea['name']
    results =[]
    for key,value in idea.items():
        if key == 'name':
            continue

        response = None
        question = neutral.format(website_idea=value)
        if context:
            response = completion(
                model=model,
                messages=[{'content':system_prompt_generator_with_retrieval,'role':'assistant'},{ "content": user_prompt_with_context.format(search_results=context,question=question),"role": "user"}]
            )
        else:
            response = completion(
                model=model,
                messages=[{'content':system_prompt_generator,'role':'assistant'},{ "content": question,"role": "user"}]
            )
        result = response.choices[0].message.content
        results.append(result)
        html = result.split('```html')[1].split('```')[0]
        path = f'websites/model/{name}'
        os.makedirs(path,exist_ok=True)
        with open(f'{path}/{key}.html','w') as file:
            file.write(html)
        print(f'{path}/{key}.html created')
    return response.choices[0].message
````

````python
def build_taxonomy_string(df :pd.DataFrame) -> str:
    return '\n\n'.join([f'### {row.Tactic}\n{row.Description}' for _, row in df.iterrows()])
def build_article(article :str, dark_patterns_taxonomy :pd.DataFrame) -> str:
    return article.format(dark_patterns_taxonomy=build_taxonomy_string(dark_patterns_taxonomy))
    
````
